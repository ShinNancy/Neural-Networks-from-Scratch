### Coding First Neuron: 3 inputs

```python
inputs = [1,2,3]
weights = [0.2,0.9, -0.5]
bias = 2

outputs = (inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias)
print(outputs)
```

### Coding First Layer

```python
input_node = [1,2,3,4]
weights =[
    [0.2, 0.3,0.4, -0.5],
    [0.3, 0.4,0.5, 0.7],
    [0.2,0.5,0.6,-0.9]
]
weightn1 = weights[0]
weightn2 = weights[1]
weightn3 = weights[2]

biasn = [2,3,0.5]

output_neurons = [
    #Neuron 1:
    input_node[0]*weightn1[0] +
    input_node[1]*weightn1[1] + 
    input_node[2]*weightn1[2] + 
    input_node[3]*weightn1[3] +
    biasn[0],

    #Neuron 2:
    input_node[0]*weightn2[0] +
    input_node[1]*weightn2[1] + 
    input_node[2]*weightn2[2] + 
    input_node[3]*weightn2[3] +
    biasn[1],

    #Neuron 3:
    input_node[0]*weightn3[0] +
    input_node[1]*weightn3[1] + 
    input_node[2]*weightn3[2] + 
    input_node[3]*weightn3[3] +
    biasn[2]
]

print(output_neurons)

```

```python
input_node = [1,2,3,4]
weights =[
    [0.2, 0.3,0.4, -0.5],
    [0.3, 0.4,0.5, 0.7],
    [0.2,0.5,0.6,-0.9]
]
biasn = [2,3,0.5]

layer_output = []

for neuron_weights, neuron_bias in zip(weights, biasn):
    neuron_output = 0
    for n_input, weight in zip(input_node, neuron_weights):
        neuron_output += n_input*weight
    neuron_output +=neuron_bias
    layer_output.append(neuron_output)

print(layer_output)
```
