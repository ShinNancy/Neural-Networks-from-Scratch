```python
# Using nnfs with vertica_data

from nnfs.datasets import vertical_data
import numpy as np
import nnfs
import matplotlib.pyplot as plt

nnfs.init()
X,y = vertical_data(samples= 100, classes= 4)
plt.scatter(X[:,0], X[:,1])
plt.show
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='brg')
plt.show
```
