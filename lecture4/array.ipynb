```python
import numpy as np
```

```python
A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
print(np.sum(A))
```

```python
help(np.sum)
```

```python
# Sum column
print(np.sum((A),axis=0))
```

```python
# Sum row
print(np.sum((A),axis=1))
```

```python
print(np.sum((A),axis=-1))
```

```python
print(np.sum(A,axis=0).shape)
print(np.sum(A,axis=0, keepdims=True).shape)
```
